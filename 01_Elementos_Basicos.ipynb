# 1. Elementos Básicos de Python

**¡Bienvenido al inicio de tu aventura con Python!** En este primer cuaderno, aprenderás los fundamentos esenciales que forman la base de todo programa. Cubriremos cómo mostrar información, guardar datos en variables, los diferentes tipos de datos y las operaciones que puedes realizar con ellos.

## La Función `print()`
La función `print()` es tu primera herramienta para comunicarte con tu programa. Te permite mostrar texto y los valores de las variables en la pantalla.

```python
# Imprimir un saludo
print("Hola, Python!")

# Imprimir números y resultados de operaciones
print(123)
print(10 + 5) # Imprime el resultado de la suma
```

## Variables
Una variable es como una caja con una etiqueta donde guardas un dato. Puedes cambiar el contenido de la caja cuando quieras.

```python
# Asignamos un valor a una variable
saludo = "Hola, soy una variable."
print(saludo)

# Podemos cambiar el valor de la variable
saludo = "He cambiado de valor."
print(saludo)

numero = 25
print(numero)
```

## Tipos de Datos
Python maneja diferentes tipos de datos. Los más comunes son:
*   **`str` (String):** Texto.
*   **`int` (Integer):** Números enteros.
*   **`float`:** Números con decimales.
*   **`bool` (Boolean):** Representa `True` (verdadero) o `False` (falso).

```python
texto = "Soy un string"
entero = 100
decimal = 99.9
es_real = True

# La función type() nos dice qué tipo de dato es una variable
print(type(texto))
print(type(entero))
print(type(decimal))
print(type(es_real))
```

## Operadores
Son símbolos que realizan operaciones sobre las variables y sus valores.

### Operadores Aritméticos

```python
a = 10
b = 3
print(f'Suma: {a + b}')
print(f'Resta: {a - b}')
print(f'Multiplicación: {a * b}')
print(f'División: {a / b}')
print(f'División Entera: {a // b}')
print(f'Módulo (resto): {a % b}')
print(f'Potencia: {a ** b}')
```

### Operadores de Comparación
Comparan dos valores y devuelven un booleano (`True` o `False`).

```python
x = 5
y = 10
print(f'¿{x} es igual a {y}? {x == y}')
print(f'¿{x} es diferente de {y}? {x != y}')
print(f'¿{x} es mayor que {y}? {x > y}')
print(f'¿{x} es menor que {y}? {x < y}')
```

### Operadores Lógicos
Combinan expresiones booleanas.

```python
p = True
q = False
print(f'{p} and {q}: {p and q}') # Y lógico (Ambas tienen que ser "True" para ser verdaderos)
print(f'{p} or {q}: {p or q}')   # O lógico (Solo una de las dos tienen que ser "True" para ser verdadero)
print(f'not {p}: {not p}')       # Negación
```

## Manipulación de Strings
Los strings vienen con métodos muy útiles para manipular texto.

```python
frase = "  Python es Divertido  "
print(frase.upper()) # Mayúsculas
print(frase.lower()) # Minúsculas
print(frase.strip()) # Sin espacios al inicio/final
print(frase.replace('Divertido', 'Genial')) # Reemplazar texto
```
