# 4. Bucles: Repitiendo Tareas

**¡Pongamos a Python a trabajar por nosotros!** Los bucles son una de las herramientas más potentes en programación. Nos permiten ejecutar un bloque de código múltiples veces sin tener que reescribirlo. En este cuaderno veremos los dos tipos de bucles de Python: `for` y `while`.

## El Bucle `for`
El bucle `for` se utiliza para **iterar sobre una secuencia** de elementos (como una lista, una tupla, un diccionario, un conjunto o un string).

```python
# Iterar sobre una lista
frutas = ["manzana", "banana", "cereza"]
print("Mis frutas favoritas son:")
for fruta in frutas:
    print(f"- {fruta}")

# Iterar sobre un string
print("Letras en la palabra 'Python':")
for letra in "Python":
    print(letra)
```

### La Función `range()`
Es muy común usar `for` con `range()` para ejecutar un bloque de código un número específico de veces.

```python
# Imprimir números del 0 al 4
print("Contando del 0 al 4:")
for i in range(5):
    print(i)

# Imprimir números del 2 al 5
print("Contando del 2 al 5:")
for i in range(2, 6):
    print(i)
```

## El Bucle `while`
El bucle `while` (mientras) ejecuta un bloque de código **mientras una condición sea `True`**.

```python
contador = 0

print("Iniciando cuenta atrás con while:")
while contador < 5:
    print(f"El contador es {contador}")
    # ¡Importante! Debemos modificar la variable de la condición
    # para evitar un bucle infinito.
    contador = contador + 1 # o contador += 1

print("¡Bucle while terminado!")
```

## `break` y `continue`
Son dos sentencias que nos dan más control sobre el comportamiento de los bucles.

### `break`
La sentencia `break` **termina el bucle inmediatamente**, sin importar la condición.

```python
# Encontrar un número y detenerse
numeros = [1, 5, 12, 25, 30, 42]
for num in numeros:
    if num == 25:
        print(f"¡Número 25 encontrado! Deteniendo el bucle.")
        break # Salimos del bucle
    print(f"Revisando el número {num}...")
```

### `continue`
La sentencia `continue` **salta a la siguiente iteración del bucle**, ignorando el resto del código de la iteración actual.

```python
# Imprimir solo los números impares
for i in range(10):
    if i % 2 == 0: # Si el número es par
        continue   # Saltamos a la siguiente iteración
    print(f"Número impar: {i}")
```
