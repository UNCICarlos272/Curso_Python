# 5. Colecciones : Listas y Tuplas

**¡Vamos a organizar nuestros datos!** A medida que tus programas crecen, necesitas formas eficientes de guardar y gestionar grupos de datos. Las colecciones son la respuesta. En este cuaderno, exploraremos las dos colecciones ordenadas más importantes: las listas y las tuplas.

## Listas (`list`)
Una lista es una colección **ordenada** y **mutable** (modificable) de elementos. Es una de las estructuras de datos más versátiles y usadas en Python.

```python
# Crear una lista
numeros = [1, 2, 3, 4, 5]
mixta = ["texto", 10, True, 3.14]
print(f"Lista de números: {numeros}")
print(f"Lista mixta: {mixta}")
```

### Acceder y Modificar Elementos
Accedemos a los elementos de una lista usando su **índice**, que empieza en 0.

```python
planetas = ["Mercurio", "Venus", "Tierra", "Marte"]

# Acceder al primer elemento (índice 0)
print(f"El primer planeta es: {planetas[0]}")

# Acceder al último elemento (índice -1)
print(f"El último planeta es: {planetas[-1]}")

# Modificar un elemento
print(f"
Lista original: {planetas}")
planetas[3] = "Jupiter" # Marte fue reemplazado
print(f"Lista modificada: {planetas}")
```

### Métodos Útiles de las Listas

```python
mi_lista = [30, 10, 40, 20]
print(f"Lista inicial: {mi_lista}")

# Añadir un elemento al final
mi_lista.append(50)
print(f"Después de append(50): {mi_lista}")

# Eliminar el último elemento
elemento_eliminado = mi_lista.pop()
print(f"Elemento eliminado con pop(): {elemento_eliminado}")
print(f"Lista después de pop(): {mi_lista}")

# Ordenar la lista
mi_lista.sort()
print(f"Lista ordenada: {mi_lista}")

# Invertir el orden de la lista
mi_lista.reverse()
print(f"Lista invertida: {mi_lista}")

# Longitud de la lista
print(f"La lista tiene {len(mi_lista)} elementos.")
```

## Tuplas (`tuple`)
Una tupla es una colección **ordenada** e **inmutable**. Una vez que creas una tupla, no puedes modificar, añadir o eliminar sus elementos. Son útiles para datos que no deben cambiar, como coordenadas o constantes.

```python
# Crear una tupla (se usan paréntesis)
coordenadas = (10.5, 45.2)
print(f"Tupla de coordenadas: {coordenadas}")

# Acceder a sus elementos (igual que las listas)
latitud = coordenadas[0]
print(f"La latitud es: {latitud}")

# Intentar modificar una tupla dará un error
# coordenadas[0] = 5.0 # Descomenta esta línea para ver el TypeError
```
